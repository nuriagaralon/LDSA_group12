```python
import matplotlib.pyplot as plt
import csv

def extract_data(file_open,data_size):
    io_data = []
    exec_data = []
    total_time = []
    number_of_cores = [1,2,4,8,16]
    for line_number,line in enumerate(file_open):
        if line_number == 0:
            index_num = line.index(data_size)
        if line_number > 1:
            exec_data.append(float(line[index_num]))
            io_data.append(float(line[index_num + 1]))
            total_time.append(float(line[index_num])+float(line[index_num+1]))
    return (io_data,exec_data,total_time,number_of_cores)

def plot_data(file_open,data_size="",weak_scaling=False):
    plt.figure()
    
    if weak_scaling:
        io,exe,total,noc = weak_scale_plot(file_open)
        plt.title("weak scaling")
        plt.text(0.75,exe[0]+25,'500mb')
        plt.text(2,exe[1]+25,'1gb')
        plt.text(4,exe[2]+25,'2gb')
        plt.text(8,exe[3]+25,'4gb')
    else:
        io,exe,total,noc = extract_data(file_open,data_size)
        plt.title("data size = " + data_size)
        
    plt.xlabel('Number of cores')
    plt.ylabel('Execution time (s)') 
    plt.plot(noc,io,"k",color="orange",label="IO time")
    plt.plot(noc,io,"bo",color="orange")
    plt.plot(noc,exe,"bo",color="green")
    plt.plot(noc,total,"bo",color="black")
    plt.plot(noc,exe,"k",color="green",label = "Execution time")
    plt.plot(noc,total,"k",color="black",label = "Total time")
        
        
    plt.legend(loc="upper right")
    if weak_scaling:
        plt.savefig("weak_scaling.jpeg",quality=95,dpi=500)
    else:
        plt.savefig(data_size+".jpeg",quality=95,dpi=500)
    plt.show()
    
def weak_scale_plot(file_open):
    weak_io_data = []
    weak_exe_data = []
    weak_total = []
    number_of_cores = [1,2,4,8]
    
    n = 0
    for line_number,line in enumerate(file_open):
        if line_number > 1 and n + 2 < len(line):
            weak_exe_data.append(float(line[1 + n]))
            weak_io_data.append(float(line[2 + n]))
            weak_total.append(float(line[1 + n])+float(line[2 + n]))
            n += 2
    return (weak_io_data,weak_exe_data,weak_total,number_of_cores)
            
with open("time_data.csv","r") as csv_file:
    file_open = list(csv.reader(csv_file,delimiter=","))
    plot_data(file_open,"500mb")
    plot_data(file_open,"1gb")
    plot_data(file_open,"2gb")
    plot_data(file_open,"4gb")
    plot_data(file_open,weak_scaling=True)
```
